```python
df = pd.read_csv('train.csv')
```

```python
df = df[:10].copy()
```

```python
import scipy
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('bert-base-nli-mean-tokens')
```

```python
question1 = 'What is the step by step guide to invest in share market in india?'
question2 = 'What is the step by step guide to invest in share market?'

question1_embeddings = model.encode(question1)
question2_embeddings = model.encode(question2)
distances = scipy.spatial.distance.cdist([question1_embeddings], [question2_embeddings], "cosine")[0][0]
print(distances)
```

```python
question1_embeddings.shape
```

```python
df.drop(['BERTCosdistance'],axis = 1,inplace = True)
```

```python
df['BERTCosdistance'] = df.apply(lambda x:scipy.spatial.distance.cdist([model.encode(x['question1'])], [model.encode(x['question2'])], "cosine")[0][0],axis = 1)
```
